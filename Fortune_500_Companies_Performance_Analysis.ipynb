```
df_companylist[['Process1', 'Process3', 'Process5']] = df_companylist[['Process1', 'Process3', 'Process5']].fillna(0)

mapping = {
    'Fortune 500': 1,
    'World’s Most Admired Companies': 1,
    'Change the World': 1,
    'The 100 Best Companies to Work For': 1,
    '100 Fastest-Growing Companies': 1
}
df_companylist = df_companylist.applymap(lambda x: mapping.get(x) if x in mapping else x)
df_companylist['No of Honors'] = df_companylist['Process1'] + df_companylist['Process3'] + df_companylist['Process5'] + int(1)
df_companylist.head()
```

```
#use seaborn plot numbers of honors each company earn
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks")
ax = sns.barplot(x = df_companylist['No of Honors'].value_counts().index, y = df_companylist['No of Honors'].value_counts())

def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        patch.set_width(new_value)

        patch.set_x(patch.get_x() + diff * .45)

change_width(ax, .5)
ax.set_title('')
ax.set_xlabel('Number of Honors', fontsize = 14)
ax.set_ylabel('Number of Companies', fontsize = 14)

plt.show()
```

```
#use matplotlib plot revenue and profit by sector
import matplotlib.pyplot as plt

revenue_by_sector = df_companylist.groupby('Sector')['Revenue($M)'].sum()
profit_by_sector = df_companylist.groupby('Sector')['Profits($M)'].sum()

fig, ax1 = plt.subplots(figsize = (20, 8))
ax2 = ax1.twinx()
revenue_by_sector.plot(kind = 'bar', color ='cornflowerblue', ax = ax1, label = 'Revenue($M)')
profit_by_sector.plot(kind = 'line', color = 'coral', ax = ax2, label = 'Profit($M)', linewidth = 5,  marker = 'o', markersize = 10)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation = 45, ha = 'right')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc = 0)
plt.title('Revenue by Sector & Profit by Sector', fontsize = 20)
ax1.xaxis.label.set_visible(False)
plt.show()
```

```
#use matplotlib plot revenue and employee by sector
employee_by_sector = df_companylist.groupby('Sector')['Employees'].sum()

fig, ax1 = plt.subplots(figsize = (20, 8))
ax2 = ax1.twinx()
revenue_by_sector.plot(kind = 'bar', color ='cornflowerblue', ax = ax1, label = 'Revenue($M)')
employee_by_sector.plot(kind = 'line', color = 'r', ax = ax2, label = 'Employee', linewidth = 5,  marker = 'o', markersize = 10)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation = 45, ha = 'right')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc = 0)
plt.title('Revenue by Sector & Employee by Sector', fontsize = 20)
ax1.xaxis.label.set_visible(False)
plt.show()
```

```
#use sklearn(Linear Regression) find relationship between revenue and employee
from sklearn import linear_model

X = df_companylist[['Revenue($M)']]
y = df_companylist['Employees'].values

model = linear_model.LinearRegression(fit_intercept = False)
res = model.fit(X, y)
plt.rcParams['figure.figsize'] = (20, 10)
plt.plot(X['Revenue($M)'], y, '.')
plt.plot(X['Revenue($M)'], model.predict(X), '.')
plt.legend(['actual employee', 'predicted employee'])
plt.show()
```

```
#use sklearn(Random Forrest Regression) find relationship between revenue and employee
from sklearn import ensemble
ensemble_model = ensemble.RandomForestRegressor()
ensemble_model.fit(X, y)

plt.plot(X[['Revenue($M)']], y, 'r.')
plt.plot(X[['Revenue($M)']], ensemble_model.predict(X), 'g.')
plt.legend(['actual employee', 'predicted employee'])
plt.show()
```

```
# Visualising the Random Forest Regression results

# arange for creating a range of values
# from min value of x to max
# value of x with a difference of 0.01
# between two consecutive values
X_grid = np.arange(min(x), max(x), 0.01)

# reshape for reshaping the data into a len(X_grid)*1 array,
# i.e. to make a column out of the X_grid value
X_grid = X_grid.reshape((len(X_grid), 1))

# Scatter plot for original data
plt.scatter(x, y, color = 'blue')

# plot predicted data
plt.plot(X_grid, regressor.predict(X_grid),
         color = 'green')
plt.title('Random Forest Regression')
plt.xlabel('Revenue change')
plt.ylabel('Revenue($M)')
plt.show()
```
